```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings("ignore")

sns.set(style="whitegrid")
```

```python
# Importanto treino e Test
df_train = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
# dropar colunas que não estão no dataset de test
columns = df_test.columns

# adicionando o target (y)
columns = columns.append(pd.Index(['NU_NOTA_MT'])) 

# ajustando apenas colunas para o modelo
df_train = df_train[columns]

#shape comparando treino e test
print('O shape de treinamento é de', df_train.shape[1], 'colunas e', df_train.shape[0], 'linhas.')
print('O shape de teste é', df_test.shape[1], 'colunas e', df_test.shape[0], 'linhas.')
```

```python
# Ajustando instrings para numeros inteiros
dict_ajust_sexo = {
    'M': 1,
    'F': 2
}

dict_ajust_q = {
    'A': 1,
    'B': 2, 
    'C': 3,
    'D': 4,
    'E': 26,
    'F': 5,
    'G': 6,
    'H': 7,
    'I': 8,
    'J': 9,
    'K': 10,
    'L': 11,
    'M': 12,
    'N': 13,
    'O': 14,
    'P': 15,
    'Q': 16,
    'R': 17,
    'S': 18,
    'T': 19,
    'U': 20,
    'V': 21,
    'W': 22,
    'X': 23,
    'Y': 24,
    'Z': 25,
}

#ajustando perguntas
# treino
df_train['Q001'].replace(dict_ajust_q, inplace=True)
df_train['Q002'].replace(dict_ajust_q, inplace=True)
df_train['Q006'].replace(dict_ajust_q, inplace=True)
df_train['Q024'].replace(dict_ajust_q, inplace=True)
df_train['Q025'].replace(dict_ajust_q, inplace=True)
df_train['Q026'].replace(dict_ajust_q, inplace=True)
df_train['Q027'].replace(dict_ajust_q, inplace=True)
df_train['Q047'].replace(dict_ajust_q, inplace=True)

#teste
df_test['Q001'].replace(dict_ajust_q, inplace=True)
df_test['Q002'].replace(dict_ajust_q, inplace=True)
df_test['Q006'].replace(dict_ajust_q, inplace=True)
df_test['Q024'].replace(dict_ajust_q, inplace=True)
df_test['Q025'].replace(dict_ajust_q, inplace=True)
df_test['Q026'].replace(dict_ajust_q, inplace=True)
df_test['Q027'].replace(dict_ajust_q, inplace=True)
df_test['Q047'].replace(dict_ajust_q, inplace=True)


# ajustando feature de sexo
df_train['TP_SEXO'].replace(dict_ajust_sexo, inplace=True)
df_test['TP_SEXO'].replace(dict_ajust_sexo, inplace=True)
```

```python
# analisando sexo
count_masculino = df_train['TP_SEXO'][df_train['TP_SEXO'] == 1].count()
count_feminino = df_train['TP_SEXO'][df_train['TP_SEXO'] == 2].count()

print('A quantidade de respondentes masculino é de:', count_masculino)
print('A quantidade de respondentes femininos é de:', count_feminino)

sex_count = df_train.groupby('TP_SEXO').count()
sex_count = sex_count[['NU_INSCRICAO']]

# plotando sexo
# 1 = masculino
# 2 = feminino
ax = sns.barplot(x=sex_count.index, y='NU_INSCRICAO', data=sex_count)
ax.set(xlabel='Sex', ylabel='total')
plt.show()
```

```python
# selecionando features
columns_final_train = ['TP_ESCOLA', 'TP_STATUS_REDACAO', 'Q001', 'Q002',
                       'NU_NOTA_REDACAO', 'NU_IDADE', 'NU_NOTA_CN',
                       'NU_NOTA_CH', 'NU_NOTA_LC', 'TP_STATUS_REDACAO',
                       'Q006', 'Q024', 'Q025', 'Q026', 'Q027', 'Q047', 'NU_NOTA_MT']

columns_final_test =  ['TP_ESCOLA', 'TP_STATUS_REDACAO', 'Q001', 'Q002',
                       'NU_NOTA_REDACAO', 'NU_IDADE', 'NU_NOTA_CN',
                       'NU_NOTA_CH', 'NU_NOTA_LC', 'TP_STATUS_REDACAO',
                       'Q006', 'Q024', 'Q025', 'Q026', 'Q027', 'Q047']

# dropando outras colunas e deixando apenas as que vamos utilizar
df_train = df_train[columns_final_train]
df_test = df_test[columns_final_test]
```

```python
# Listando missing values
# tp_ensino
# nu_nota_cn
# nu_nota_ch
# tp_status_redacao
# nu_nota_redacao
# q026
# nu_nota_mt

# printando colunas nulas de cada dataset
null_counts_train = df_train.isnull().sum()
print('Missing Values Dataset - Treino \n')
print(null_counts_train[null_counts_train > 0].sort_values(ascending=False), '\n')

print('--------------------------- \n')
print('Missing Values Dataset - Test \n')

null_counts_test= df_test.isnull().sum()
print(null_counts_test[null_counts_test > 0].sort_values(ascending=False))
```

```python
# 0: não informado
# N: não informado
df_train.fillna({
                 'TP_ENSINO': 0,
                 'Q027': 0,
                 'NU_NOTA_REDACAO': 0,
                 'TP_STATUS_REDACAO': 0,
                 'NU_NOTA_LC': 0,
                 'NU_NOTA_CH': 0,
                 'NU_NOTA_CN': 0}, inplace=True)

df_test.fillna({
                 'TP_ENSINO': 0,
                 'Q027': 'N',
                 'NU_NOTA_REDACAO': 0,
                 'TP_STATUS_REDACAO': 0,
                 'NU_NOTA_LC': 0,
                 'NU_NOTA_CH': 0,
                 'NU_NOTA_CN': 0}, inplace=True)

# retirando os nulls do target
df_train.dropna(subset=['NU_NOTA_MT'], inplace=True)

# testando alteração
# printando colunas nulas de cada dataset
null_counts_train = df_train.isnull().sum()
print('Missing Values Dataset - Treino \n')
print(null_counts_train[null_counts_train > 0].sort_values(ascending=False), '\n')

print('--------------------------- \n')
print('Missing Values Dataset - Test \n')

null_counts_test= df_test.isnull().sum()
print(null_counts_test[null_counts_test > 0].sort_values(ascending=False))
```

```python
# correlação dos dados, features com target
corr = df_train.corr()
corr.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

```python
# alterando tipos para int
# treinO
df_train['TP_STATUS_REDACAO'] = df_train['TP_STATUS_REDACAO'].astype(int)


# test
df_test['TP_STATUS_REDACAO'] = df_test['TP_STATUS_REDACAO'].astype(int)
```

```python
x_train = df_train.drop("NU_NOTA_MT", axis=1)
y_train = df_train["NU_NOTA_MT"] 

x_test = df_test.drop("NU_INSCRICAO", axis=1)
```

```python
# Algorithms para treinar o modelo
from sklearn import linear_model
from sklearn.model_selection import cross_val_score, train_test_split, KFold
from sklearn.linear_model import LinearRegression, LassoCV, LassoLarsCV , Ridge
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
print('O shape de x_train é:', x_train.shape)
print('O shape de y_train é:', y_train.shape)
print('O shape de x_test é:', x_test.shape)
```

```python
# reg
reg = linear_model.LinearRegression()
reg.fit(x_train, y_train)
#y_reg_pred = reg.predict(x_test)

seed = 7
r2 = 'r2'
mse = 'neg_mean_squared_error'
mae = 'neg_mean_absolute_error'

kfold = KFold(n_splits=10, random_state=seed)

results = cross_val_score(reg, x_train, y_train, cv=kfold, scoring=r2)
print("R^2:", round(results.mean(),3) , '(',round(results.std(), 3), ')')

results = cross_val_score(reg, x_train, y_train, cv=kfold, scoring=mse)
print("MSE:", round(results.mean(),3) , '(',round(results.std(), 3), ')')

results = cross_val_score(reg, x_train, y_train, cv=kfold, scoring=mae)
print("MAE:", round(results.mean(),3) , '(',round(results.std(), 3), ')')
```

```python
# Ridge
rdg = Ridge()
rdg.fit(x_train, y_train)
#y_rdg_pred = rdg.predict(x_test_1)

results = cross_val_score(rdg, x_train, y_train, cv=kfold, scoring=r2)
print("R^2:", round(results.mean(),3) , '(',round(results.std(), 3), ')')

results = cross_val_score(rdg, x_train, y_train, cv=kfold, scoring=mse)
print("MSE:", round(results.mean(),3) , '(',round(results.std(), 3), ')')

results = cross_val_score(rdg, x_train, y_train, cv=kfold, scoring=mae)
print("MAE:", round(results.mean(),3) , '(',round(results.std(), 3), ')')
```

```python
def rmse_cv(model):
    rmse= np.sqrt(cross_val_score(model, x_train, y_train, scoring="neg_mean_squared_error", cv = 5))
```

```python
model_lasso = LassoCV(alphas = [1, 0.1, 0.001, 0.0005]).fit(x_train, y_train)
```

```python
results = cross_val_score(model_lasso, x_train, y_train, cv=kfold, scoring=mse)
print("MSE:", round(results.mean(),3) , '(',round(results.std(), 3), ')')
```

```python
coef = pd.Series(model_lasso.coef_, index = x_train.columns)
```

```python
print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " +  str(sum(coef == 0)) + " variables")
```

```python
imp_coef = pd.concat([coef.sort_values().head(10),
                     coef.sort_values().tail(10)])
```
